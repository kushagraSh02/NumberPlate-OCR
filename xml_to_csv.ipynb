```python
import pandas as pd
import xml.etree.ElementTree as xet
from glob import glob
```

```python
path = glob('./data/*.xml')
path
```

```python
file = path[0]
info = xet.parse(file)
root = info.getroot()
```

```python
member_obj = root.find('object')
labels_info = member_obj.find('bndbox')
xmin = int(labels_info.find('xmin').text)
xmax = int(labels_info.find('xmax').text)
ymin = int(labels_info.find('ymin').text)
ymax = int(labels_info.find('ymax').text)
print(xmin, xmax, ymin, ymax)
```

```python
label_dict = dict(filepath=[], xmin=[], xmax=[], ymin=[], ymax=[])
for filename in path:
    info = xet.parse(filename)
    root = info.getroot()
    member_obj = root.find('object')
    labels_info = member_obj.find('bndbox')
    xmin = int(labels_info.find('xmin').text)
    xmax = int(labels_info.find('xmax').text)
    ymin = int(labels_info.find('ymin').text)
    ymax = int(labels_info.find('ymax').text)
    label_dict['filepath'].append(filename)
    label_dict['xmin'].append(xmin)
    label_dict['xmax'].append(xmax)
    label_dict['ymin'].append(ymin)
    label_dict['ymax'].append(ymax)
```

```python
label_dict
```

```python
df = pd.DataFrame(label_dict)
df
```

```python
df.to_csv('labels.csv', index=False)
```
